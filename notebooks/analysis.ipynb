```python
import sys
sys.path.append('../') # Access the src folder
from src.covid_engine import CovidAnalyticsEngine
import matplotlib.pyplot as plt

# Initialize Engine
engine = CovidAnalyticsEngine('../data/raw/time-series-19-covid-combined.csv')

# Get data for a specific country
df_plot = engine.get_velocity_metrics("US")

# Plotting the raw data vs. our smoothed NumPy average
plt.figure(figsize=(12,6))
plt.bar(df_plot['Date'], df_plot['Daily_New'], alpha=0.3, label='Daily New Cases (Raw)')
plt.plot(df_plot['Date'], df_plot['7Day_MA'], color='red', label='7-Day Moving Average (Smoothed)')
plt.title("Infection Velocity - Vectorized Analysis")
plt.legend()
plt.show()
```
